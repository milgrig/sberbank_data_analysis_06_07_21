### Упражнение 1

* Скачайте с портала Открытых Данных производственный календарь России: https://data.gov.ru/opendata/7708660670-proizvcalendar
* На его основе определите соответствующий календарь в Pandas: `RussianBusinessCalendar(AbstractHolidayCalendar)`
* С помощью полученного в пункте выше календаря и модуля `pd.offsets` создайте DataFrame one-hot календарных признаков:
    * День недели
    * Месяц
    * Выходной день
    * Праздничный день
    * Предпраздничный рабочий день
    * Последний день месяца
    * Последний рабочий день месяца
    * Предновогодний выходной день

```python
import datetime # для работы со временем 

import numpy as np # уже знакомые библиотеки
import pandas as pd
import matplotlib.pyplot as plt
from pandas.tseries.holiday import AbstractHolidayCalendar, Holiday, nearest_workday, EasterMonday
from pandas.tseries.offsets import DateOffset, Day

%matplotlib inline
```

```python
calendar = pd.read_csv('data.csv')
calendar
```

```python
r1 = [] # выходные дни/праздничные дни
r2 = [] # предпраздничные дни
r3 = [] # перенесённый выходной день
for c in calendar.index:
    for m in range(1,13):
        z = calendar[calendar.columns[m]][c]
        z = z.replace(',', ' ')
        z = z.split()
        r1 += [datetime.datetime(calendar['Год/Месяц'][c], m, int(i)) for i in z if  i.isdecimal()]
        r2 += [datetime.datetime(calendar['Год/Месяц'][c], m, int(i.replace('*', '')))  for i in z if  i.find('*') > 0]
        r3 += [datetime.datetime(calendar['Год/Месяц'][c], m, int(i.replace('+', '')))  for i in z if  i.find('+') > 0]
```

```python
start_date = datetime.datetime(1999, 1, 1)
end_date = datetime.datetime(2025, 12, 31)
class RussianBusinessCalendar(AbstractHolidayCalendar):
    start_date = datetime.datetime(1999, 1, 1)
    end_date = datetime.datetime(2025, 12, 31)
    rules = [Holiday('Рабочая суббота', month=3, day=8, year=2020, offset=Day(1))]

russian_busday = pd.offsets.CustomBusinessDay(holidays = r1, weekmask='Mon Tue Wed Thu Fri' ,
                                              calendar=RussianBusinessCalendar())
pre_holiday = pd.offsets.CustomBusinessDay(holidays = r2, weekmask='Mon Tue Wed Thu Fri' ,
                                              calendar=RussianBusinessCalendar())
```

```python
r4 = [datetime.datetime(2026, 12, 1) - russian_busday] #последний рабочий день месяца
r5 = [datetime.datetime(2026, 12, 1) - 2*russian_busday] # предпоследний рабочий день месяца
for y in range(1999,2026):
    for m in range(1,13):
        r4 += [datetime.datetime(y, m, 1) -russian_busday]
        r5 += [datetime.datetime(y, m, 1) -2*russian_busday]
```

```python
r6 = pd.date_range(start='2000-1-2', end='2000-1-1', freq='D') #пустой date_range, добавим в него предновогодние выходные
for y in range(2000,2027):
    st = datetime.datetime(y, 1, 1) - russian_busday + pd.Timedelta('1D')
    en = datetime.datetime(y, 1, 1) - pd.Timedelta('1D')
    r6 = r6.union(pd.date_range(start=st, end=en, freq='D'))

    
```

```python
columns=['Выходной','Предпраздничный рабочий день']
week = ['Пн', 'Вт', 'Ср', 'Чт', 'Пт','Сб','Вс']
months = ['Январь', 'Февраль', 'Март', 'Апрель', 'Май', 'Июнь', 'Июль', 'Август', 'Сентябрь', 'Октябрь', 'Ноябрь', 'Декабрь']
ts = pd.date_range(start='1999-01-01', end='2025-12-31', freq='D')
df = pd.DataFrame(index=ts, columns=columns)
df = df.fillna(0)
df.loc[df.index.isin(russian_busday.holidays),'Выходной']=1
df.loc[df.index.isin(pre_holiday.holidays),'Предпраздничный рабочий день']=1
df['Последний рабочий день мессяца'] = [int(i in r4) for i in df.index] 
df['Предпоследний рабочий день мессяца'] = [int(i in r5) for i in df.index] 
df['weekday'] = [i.weekday() for i in df.index]
for i in range(7):
    df[week[i]] = 0
    df.loc[df['weekday'] == i, week[i]] = 1
df.drop(['weekday'], axis = 1, inplace = True)
for i in range(12):
    df[months[i]] = 0
    df.loc[df.index.month == i+1, months[i]] = 1   
df['Предновогодний выходной день'] = [int(i in r6) for i in df.index]     
    
```

### Упражнение 2

* Для временного ряда курса акций Сбербанка на основе производственного календаря России сформируйте DataFrame признаков:

    * Статистики `['mean', 'median', 'max', 'min', 'std'] ` в расширяющемся окне сгруппированные по рабочим, праздничным и предпраздничным дням

```python
ts = pd.read_csv('../data/sber_stocks.csv', usecols=['CLOSE', 'TRADEDATE'], 
                 index_col='TRADEDATE', parse_dates=True).CLOSE
#ts.plot(figsize=(20, 10), grid=True, linewidth=3)
```

Лаговые переменные рабочих дней с глубинами: 1, 3, 5

```python
l1 = pd.DataFrame(ts.shift(1, freq=russian_busday))
l3 = pd.DataFrame(ts.shift(3, freq=russian_busday))
l5 = pd.DataFrame(ts.shift(5, freq=russian_busday))
l1 = l1.merge(l3, how='outer' , left_index=True, right_index=True)
res = l1.merge(l5, how='outer' , left_index=True, right_index=True)
res.rename(columns={'CLOSE_x': 'Lag 1', 'CLOSE_y': 'Lag 3', 'CLOSE': 'Lag 5'}, inplace=True)
res
```

Скользящие статистики ['mean', 'median', 'max', 'min', 'std'] с окнами 1, 3, 5 сгрупированные по дням недели

```python
ts = pd.DataFrame(ts)
ts['weekday'] = ts.index.weekday
s1 = ts.groupby('weekday').rolling(window=1).agg([np.mean, np.median, np.max, np.min, np.std]).CLOSE
s1.rename(columns={'mean': '1 mean', 'median': '1 median', 'amax': '1 max', 'amin': '1 min', 'std': '1 std'}, inplace=True)
s3 = ts.groupby('weekday').rolling(window=3).agg([np.mean, np.median, np.max, np.min, np.std]).CLOSE
s3.rename(columns={'mean': '3 mean', 'median': '3 median', 'amax': '3 max', 'amin': '3 min', 'std': '3 std'}, inplace=True)
s5 = ts.groupby('weekday').rolling(window=5).agg([np.mean, np.median, np.max, np.min, np.std]).CLOSE
s5.rename(columns={'mean': '5 mean', 'median': '5 median', 'amax': '5 max', 'amin': '5 min', 'std': '5 std'}, inplace=True)
s1 = s1.merge(s3, how='outer' , left_index=True, right_index=True)
s1 = s1.merge(s5, how='outer' , left_index=True, right_index=True)
s1
```

Статистики ['mean', 'median', 'max', 'min', 'std'] в расширяющемся окне сгруппированные по рабочим, праздничным и предпраздничным дням

```python
df['Тип дня'] = df['Предпраздничный рабочий день'] * 2 + df['Выходной']
cal = df.replace({'Тип дня' : { 0 : 'Рабочий день', 1 : 'Предпраздничный рабочий день', 2 : 'Выходной'}})['Тип дня']
cal = pd.DataFrame(cal)
rez = ts.merge(cal,  left_index=True, right_index=True)
rez = rez.groupby('Тип дня').rolling(window=5).agg([np.mean, np.median, np.max, np.min, np.std]).CLOSE
rez.head(50)
```
